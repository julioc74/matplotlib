```python
# Importamos las librerias correspondientes
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

# Generamos una semilla aleatoria
np.random.seed(107)

# Creamos un dataframe de manera aleatoria
columnas = ["Hobbies", "Edad", "Genero"]

hobbies = np.random.choice(["musica", "deporte", "dibujo", "lectura", "paseo"], size=50)
edad = np.random.randint(18,65, size=50)
genero = np.random.choice(["M", "F"], size=50)

datos = list(zip(hobbies, edad, genero))
df = pd.DataFrame(datos, columns=columnas)
df
```

De la columna "Hobbies" contamos cuantos elementos tiene cada elemento.

```python
df_hobbies =  df["Hobbies"].value_counts()
df_hobbies
```

Convertimos a diccionario el dataframe df_hobbies y luego obtenemos una lista de 
claves y valores para usarlos como argumento en el grafico de ax.pie 

```python
dicc = dict(df_hobbies)
claves = list(dicc.keys())
valores = list(dicc.values())
```

```python
# Creamos una figura y un conjunto de ejes con plt.subplots()
fig, ax = plt.subplots()
ax.pie(valores, labels=claves, autopct='%1.2f%%', startangle=90 )


# Creamos un circulo blanco con plt.Circle y lo agregamos al grafico
# usando add_artist()
centro_circ = plt.Circle((0,0), 0.70, fc='white') 
fig = plt.gcf()
fig.gca().add_artist(centro_circ)


# Asegura que el grafico de donuts tenga una forma circular, ajustando los ejes 
# para que tengan la misma escala
ax.axis('equal')


# Ajusta automaticamente el grafico 
plt.tight_layout()

# Muestra el grafico
plt.show()
```
