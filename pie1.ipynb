```python
import pandas as pd
import matplotlib.pyplot as plt

columnas = ["Nombres", "Aficion principal", "Pais"]

nombres = ["Antonio", "Liliana", "Ana", "Daniel", "Luis", "Laura", "Julio", "Jorge", "Tania", "Cintia", "Ana", "Jose", 
           "Carlos", "Julia", "Pedro", "Ester", "Alberto", "Tatiana", "Pablo", "Marta", "Jorge", "Rosa", "Jesica", 
           "Ursula", "Luis", "Jose", "Miguel", "Raul", "Susana", "Vanesa", "Hiroshi", "Frank", "Anand", "Farah", "Jiyu",
           "Kumar", "Chang Min", "Rose", "Bastian", "Olga", "Jim"]

aficion_principal = ["Basquet", "Futbol", "Informatica", "Ajedrez", "Voley", "Futbol", "Ajedrez", "Karaoke", "Judo",
                     "Musica", "Reposteria", "Futbol", "Informatica", "Basquet", "Karaoke", "Musica", "Judo", "Teatro",
                     "Futbol", "Karaoke", "Lectura", "Futbol", "Voley", "Judo", "Ajedrez", "Voley", "Ajedrez", "Ajedrez",
                     "Karaoke", "Basquet", "Beisbol", "Basquet", "Criquet", "Voley", "Musica", "Criquet", "Basquet", 
                     "Musica", "Judo", "Ajedrez", "Karaoke"]

pais = ["Italia", "Argentina", "Cuba", "Colombia", "Venezuela", "Argentina", "Usa", "Brasil", "Peru", "Colombia", "Alemania",
         "Sudafrica", "Chile", "Bolivia", "Costa Rica" , "Panama", "Mexico", "Colombia", "Argentina", "Espaa", "Honduras", 
         "Francia", "Peru", "Mexico", "Brasil", "Ecuador", "Usa", "Mexico", "Canada", "Argentina", "Japon", "Australia",
         "India", "Iran", "Corea del sur", "India", "China", "Reino Unido", "Francia", "Rusia", "Philipinas"]

datos = list(zip(nombres, aficion_principal, pais))

df = pd.DataFrame(datos, columns=columnas)

""" De la columna "Aficion principal" del dataframe anterior con la funcion value_counts contamos cuantas personas existen 
    para cada una de las aficiones, y lo guardamos en la variable df"""
df = df["Aficion principal"].value_counts()


""" Como esta funcion "value_counts()" nos da como resultado la cantidad de personas que existe por cada aficion, esto es
    nos devuelve un par de datos que lo podemos relacionar como clave-valor, entonces lo podemos convertir en un diccionario
    y luego lo guardamos en la variable dicc. Por lo tanto podemos de ello obtener tanto su clave como su valor y los 
    podemos almacenar en las variables aficion y cantidad respectivamente"""
dicc = dict(df)

aficion = dicc.keys()
cantidad = dicc.values()

""" Finalmente podemos usar el metodo pie para obtener un grafico circular  usando los parametros x, labels y autopct, este 
     ultimo para darle formato a los valores en porcentaje mostrado en el grafico pie"""
plt.pie(x = cantidad, labels=aficion, autopct="%1.1f%%")
plt.show()
```
