```python
# Influenciado libro python 3 jap pag 336
import matplotlib.pyplot as plt

periodo1, notas_matematicas = range(0, 5), [11, 16, 14, 7, 13]
periodo2, notas_historia = range(0, 5), [8, 12, 15, 14, 11]
alumnos = ["Juan", "Maria", "Luis", "Jorge", "Ana"]

# Crea una figura de matplotlib en la que se colocarán los subplots.
fig = plt.figure()


"""Crea el primer subplot en la figura. El argumento (1, 2, 1) significa
  que habrá una fila y dos columnas de subplots, y este será el primer subplot."""
ax1 = fig.add_subplot(1, 2, 1)

"""Crea un gráfico de barras en el primer subplot. periodo1 se utiliza en 
  el eje X, notas_matematicas en el eje Y y alumnos se usan como etiquetas 
  para el eje X."""
ax1.bar(periodo1, notas_matematicas, tick_label = alumnos)

# Establece el titulo del primer grafico
ax1.set_title("Matematicas")



""" Crea el segundo subplot en la figura. El argumento (1, 2, 2) significa 
  que habrá una fila y dos columnas de subplots, y este será el segundo subplot."""
ax2 = fig.add_subplot(1, 2, 2)

"""Crea un gráfico de barras en el segundo subplot. periodo2 se utiliza en 
  el eje X, notas_historia en el eje Y y alumnos se usan como etiquetas 
  para el eje X."""
ax2.bar(periodo2, notas_historia, tick_label = alumnos)

# Establece el titulo del segundo grafico
ax2.set_title("Historia")


# Muestra los graficos
plt.show() 
```
