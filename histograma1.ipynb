```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Establecer una semilla para generar n√∫meros aleatorios reproducibles
np.random.seed(123)

# Genera datos aleatorios para cada columna
edades = np.random.randint(20, 65, size=50)
estatura = np.random.randint(150, 190, size=50)
ciudades = np.random.choice(["Tokio", "Nagoya", "Yokohama", "Osaka", "Kyoto", "Hokaido", "Okinawa"], size=50)

# Creacion del dataframe con las columnas generadas
datos = list(zip(edades, estatura, ciudades))
df = pd.DataFrame(datos, columns=["Edad",  "Estatura", "Ciudad"])
df
```

```python
intervalos_estatura =  [150, 160, 170, 180, 190]
plt.hist(x=estatura, bins=intervalos_estatura, rwidth=0.8)
plt.title("histograma de estaturas")
plt.xlabel("intervalo de estaturas")
plt.ylabel("frecuencia")
plt.show()
```
