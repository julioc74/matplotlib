# PLOT

## Generamos un dataframe de 8 productos con cantidades y precios generados al azar con el fin de obtener graficos de plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creamos una semilla para generar numeros aleatorios
np.random.seed(125)

columnas = ["Producto", "Cantidad", "Precio"]

producto = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
cantidad = np.random.randint(50, 500, size=8)
# Redondeamos a dos decimales los numeros decimales obtenidos al azar
# entre 20 y 100 
precio = np.round(np.random.uniform(20, 100, size=8), 2)

datos = list(zip(producto, cantidad, precio))

df = pd.DataFrame(datos, columns=columnas)
print(df)
```

```python
# Definir 'posiciones' como un rango que va desde 0 hasta la longitud de la lista 'producto'
posiciones = range(len(producto))

# Crear un gráfico de líneas utilizando la lista 'cantidad'
plt.plot(cantidad)

# Etiquetar el eje x del gráfico como "producto"
plt.xlabel("producto")

# Cambiamos las etiquetas en el eje x por la lista 'producto' en lugar de la lista 'posiciones'
plt.xticks(posiciones, producto)

# Agregar cuadriculas tanto en el eje x como en el eje y
plt.grid(axis='both')

# Etiquetar el eje y del gráfico como "cantidad"
plt.ylabel("cantidad")

# Mostramos el grafico
plt.show()
```

```python
plt.plot(precio)
plt.xlabel("producto")
plt.ylabel("precio")
plt.xticks(posiciones, producto)

# Agregar una cuadricula en el eje y
plt.grid(axis='y')
plt.show()
```

```python
# Crear una figura para contener los dos subgráficos
fig = plt.figure()

# Definir el primer subgráfico en una cuadrícula de 1 fila y 2 columnas, y seleccionar la primera posición
ax1 =fig.add_subplot(1, 2, 1)

# Crear un gráfico de líneas en el primer subgráfico que muestra la relación 
# entre 'producto' y 'cantidad' con marcadores de estrella
ax1.plot(producto, cantidad, marker='*')

# Etiquetar el eje x del primer subgráfico como "producto"
ax1.set_xlabel("producto")

# Etiquetar el eje y del primer subgráfico como "cantidad"
ax1.set_ylabel("cantidad")

# Agregar una cuadrícula en el eje y del primer subgráfico
ax1.grid(axis='y')



# Definir el segundo subgráfico en la misma cuadrícula, seleccionando la segunda posición
ax2 = fig.add_subplot(1, 2, 2)

# Crear un gráfico de líneas en el segundo subgráfico que muestra la relación entre 'producto' 
# y 'precio' con marcadores de estrella
ax2.plot(producto, precio, marker='*')

# Etiquetar el eje x del segundo subgráfico como "producto"
ax2.set_xlabel("producto")

# Etiquetar el eje y del segundo subgráfico como "precio"
ax2.set_ylabel("precio")

# Agregar una cuadrícula en el eje y del segundo subgráfico
ax2.grid(axis='y')



# Ajustar automaticamente el espaciado entre los dos subgráficos
plt.tight_layout(w_pad=3.0)

# Mostrar la figura que contiene ambos subgráficos
plt.show()
```
