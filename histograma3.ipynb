```python
import matplotlib.pyplot as plt
import numpy as np

# Genera 100 valores aleatorios entre 1 y 999
valores = np.random.randint(1, 1000, size=100)

# Define los límites de los intervalos
bins = np.arange(1, 1101, 100)

# Calcula el histograma automaticamente 
hist, edges = np.histogram(valores, bins=bins)

""" i = 0      i = 1       i = 2              i = 9       i = 10
      |----------|-----------|..................|-----------|
      1         101         201                901         1000 
      
      i = 0  --> edges[0] = 1  y edges[0+1] - 1 = edges[1] - 1 = 101 - 1 = 100 --> 1-100
      i = 1  --> edges[1] = 101 y edges[1+1] - 1 = edges[2] - 1 = 201 - 1 = 200 --> 101-200
      y asi suscesivamente se forman los intervalos deseados.
      
      y además nótese que se itera desde i = 0 hasta i = 9, porque edges tiene 11 elementos y queremos
      formar 10 intervalos."""

# Usando listas por comprensión generamos las etiquetas en el eje x
etiquetas = [f"{edges[i]}-{edges[i+1]-1}" for i in range(len(edges) - 1)]

plt.bar(range(len(hist)), hist, tick_label=etiquetas, width=0.9)
plt.xlabel("Intervalo de valores")
plt.ylabel("Frecuencia")
plt.title("Histograma de valores aleatorios")
plt.xticks(rotation=45)
plt.show()
```
