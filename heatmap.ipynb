```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt


# Crea una matriz de 10x10 con valores aleatorios de nmeros flotantes en el rango [0, 1].

data = np.random.rand(10, 10)

# sns.heatmap(data): Crea un mapa de calor basado en los valores de data.
# annot=True: Muestra los valores numricos dentro de cada celda. (0 ser un color ms oscuro, 1 ser ms claro). 
sns.heatmap(data, annot=True)
plt.title('Heatmap')
plt.show()
```
