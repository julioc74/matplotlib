```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt


# Crea una matriz de 10x10 con valores aleatorios de números flotantes en el rango [0, 1].

data = np.random.rand(10, 10)

# sns.heatmap(data): Crea un mapa de calor basado en los valores de data.
# annot=True: Muestra los valores numéricos dentro de cada celda. (0 será un color más oscuro, 1 será más claro). 
sns.heatmap(data, annot=True)
plt.title('Heatmap')
plt.show()
```
