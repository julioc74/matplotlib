```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generamos un numero random para la semilla
np.random.seed(123)

columnas = ["Genero", "Edad", "Estatura (cm)", "Profesion", "Ciudad"]

# Generamos valores aleatorios en cada columna
genero = np.random.choice(["M", "F"], size=50)
edad = np.random.randint(20, 65, size=50)
estatura = np.random.randint(148, 192, size=50)
profesion = np.random.choice(["Ingeniero", "Medico", "Abogado", "Economista", "Contador", 
                              "Administrador", "Profesor", "Artista"], size=50)
ciudad = np.random.choice(["Nagoya", "Osaka", "Kyoto", "Yokohama", "Tokio", 
                           "Hokaido", "Okinawa"], size=50)


# Creamos el dataframe df
datos = list(zip(genero, edad, estatura, profesion, ciudad))
df = pd.DataFrame(datos, columns=columnas)
df
```

Del dataframe original obtenemos solo el grupo de varones

```python
varones = df.groupby('Genero').get_group('M')
varones
```

Del dataframe original obtenemos solo el grupo de damas.

```python
damas = df.groupby('Genero').get_group('F')
damas
```

Hacemos uso del metodo value_counts() para contar los varones por profesion.

```python
varones['Profesion'].value_counts()
```

Ahora contamos los varones por ciudad.

```python
varones['Ciudad'].value_counts()
```

Luego contamos las damas por profesion.

```python
damas['Profesion'].value_counts()
```

Comprobando la cantidad de damas por rango de edades para contrastarlo
luego en el histograma.

```python
contador_lista = [0] * 9

for edad in damas['Edad']:
    if edad < 25:
        contador_lista[0] += 1
    elif edad < 30:
        contador_lista[1] += 1
    elif edad < 35:
        contador_lista[2] += 1
    elif edad < 40:
        contador_lista[3] += 1
    elif edad < 45:
        contador_lista[4] += 1    
    elif edad < 50:
        contador_lista[5] += 1
    elif edad < 55:
        contador_lista[6] += 1    
    elif edad < 60:
        contador_lista[7] += 1
    elif edad < 65:
        contador_lista[8] += 1      

for i in range(9):
    print(f"hay {contador_lista[i]} damas en el rango {i+1}")
```

```python

# Crear una figura con 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(15,12))

# Diagrama de barras horizontal para profesiones de varones
profesiones_v = df[df['Genero'] == 'M']['Profesion'].value_counts()
axes[0, 0].barh(profesiones_v.index, profesiones_v)
axes[0, 0].set_title("Varones por profesiones")

# Diagrama de barras para ciudades de varones
ciudades_v = df[df['Genero'] == 'M']['Ciudad'].value_counts()
axes[0, 1].bar(ciudades_v.index, ciudades_v) 
axes[0, 1].set_title("Varones por ciudad")

# Diagrama de pastel para profesiones de damas
profesiones_d = df[df['Genero'] == 'F']['Profesion'].value_counts()
axes[1, 0].pie(profesiones_d, labels=profesiones_d.index, autopct="%1.1f%%")
axes[1, 0].set_title("Damas por profesiones")

# Histograma de edades de damas
axes[1, 1].hist(damas['Edad'], bins=9, edgecolor="black")
axes[1, 1].set_title("Rango de edades en damas")
axes[1, 1].set_xlabel("edades")
axes[1, 1].set_ylabel("numero de mujeres")

# Ajustar espaciado entre subplots
plt.tight_layout(h_pad=3.0, w_pad=3.0)

# Mostrar la figura
plt.show()
```
