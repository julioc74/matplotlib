```python
import numpy as np
import matplotlib.pyplot as plt

aceite_marcas = ["Health", "Standar", "Bonito", "Class-A1", "Super", "Nature", "Color", "Sol", "Fresh", "Santa Ana", "Simple", "Rich"]
# Dadas en litros
aceite_cantidades_envase = [1.2, 1.5, 1.0, 1.2, 0.9, 1.2, 1.5, 1.0, 0.8, 1.1, 1.2, 1.5]
# Dados en $
aceite_precio = [2.55, 3.1, 1.98, 2.6, 2.0, 2.6, 3.4, 1.85, 1.7, 2.45, 2.3, 3.7]

# Uniformizando los precios a 1 litro
aceite_precio_litro = list(map(lambda x1, x2: x1/x2, aceite_precio, aceite_cantidades_envase))
aceite_precio_litro_round = list(map(lambda p : round(p, 2), aceite_precio_litro))

print(aceite_precio_litro)
# Precio de aceite por litro redondeado a dos decimales
print(aceite_precio_litro_round)

plt.xticks(rotation='vertical')
plt.xlabel("marca de aceite")
plt.ylabel("precio de aceite por litro")
plt.scatter(aceite_marcas, aceite_precio_litro_round)
plt.show()

```

```python

plt.xlabel("Marca de aceite")
plt.ylabel("precio")
plt.xticks(rotation='vertical')
plt.scatter(aceite_marcas, aceite_precio, label="precio del aceite por envase", c='orange', marker='*')
plt.scatter(aceite_marcas, aceite_precio_litro_round, label="precio del aceite por litro", c='blue')
plt.legend()
plt.show()
```
