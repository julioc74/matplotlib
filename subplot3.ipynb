Creamos un dataframe con 100 alumnos con notas de los cursos de matematicas y gramatica. 
Tanto sus notas como su genero (M o F) seran generados de manera aleatoria con la 
clase random de numpy.

Para luego generar graficos en matplotlib de boxplot, histogramas y plot de acuerdo 
a los cursos y genero a traves de subplot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(125)


notas_matematicas = np.random.randint(5, 20, size=100)
notas_gramatica = np.random.randint(5, 20, size=100)
genero = np.random.choice(['M', 'F'], size=100)

columnas = ['matematicas', 'gramatica', 'genero']
datos = list(zip(notas_matematicas, notas_gramatica, genero))

df = pd.DataFrame(datos, columns=columnas)
print(df)
```

```python
varones = df.groupby('genero').get_group('M')
damas = df.groupby('genero').get_group('F')

print(varones)
print()
print(damas)
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.boxplot(varones.matematicas)
ax1.set_title("notas de varones matematicas")
ax1.set_ylabel("notas")
ax1.grid(axis='y')

ax2 = fig.add_subplot(2,2,2)
ax2.boxplot(damas.matematicas)
ax2.set_title("notas de damas matematicas")
ax2.set_ylabel("notas")
ax2.grid(axis='y')


ax3 = fig.add_subplot(2,2,3)
ax3.boxplot(varones.gramatica)
ax3.set_title("notas de varones gramatica")
ax3.set_ylabel("notas")
ax3.grid(axis='y')


ax4 = fig.add_subplot(2,2,4)
ax4.boxplot(damas.gramatica)
ax4.set_title("notas de damas gramatica")
ax4.set_ylabel("notas")
ax4.grid(axis='y')


plt.tight_layout(w_pad=3.0, h_pad=3.0)

plt.show()

```

```python
fig = plt.figure(figsize=(12, 6))
intervalos = range(5,21)

ax1 = fig.add_subplot(1,2,1)
ax1.hist(x=notas_matematicas, bins=intervalos, rwidth=0.7 )
ax1.set_title("matematicas")
ax1.set_xlabel("notas")
ax1.set_ylabel("frecuencia")

ymin, ymax = plt.ylim()

ax2 = fig.add_subplot(1,2,2)
ax2.hist(x=notas_gramatica, bins=intervalos, rwidth=0.8)
ax2.set_title("gramatica")
ax2.set_xlabel("notas")
ax2.set_ylabel("frecuencia")

plt.ylim(ymin,ymax)
plt.tight_layout(w_pad=3.0)

plt.show()
```
