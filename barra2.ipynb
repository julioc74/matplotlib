En el siguiente diagrama de barras vemos las notas de Juan y Maria de manera
contigua. 

```python
import matplotlib.pyplot as plt
import numpy as np
juan = [14, 16, 11, 15, 14]
maria = [7, 14, 11, 13, 16]
jose = [11, 9, 15, 5, 17]
periodo = ["primer", "segundo", "tercer", "cuarto", "quinto"]

# array de 0 a 4, array([0, 1, 2, 3, 4])
indx = np.arange(len(periodo))
# ancho de cada barra
bar_width = 0.35

fig, ax = plt.subplots()
# El primer parametro de la funcion bar indica donde esta el centro
# de la barra en este caso juan y como indx es [0, 1, 2, 3, 4]
# el centro de la primera barra de juan estara en 0 - 0.35/2
# el centro de la segunda barra de juan estara en 1 - 0.35/2
# asi sucesivamente
bar_juan = ax.bar(indx -bar_width/2, juan, bar_width, label="notas de juan")

# para maria...
# el centro de la primera barra estara en 0 + 0.35/2
# el centro de la segunda barra estara en 1 + 0.35/2
# asi sucesivamente
bar_maria = ax.bar(indx + bar_width/2, maria, bar_width, label="notas de maria")

plt.legend()
plt.show()
```

Vemos la misma grafica anterior solo que reemplazamos los valores iniciales
en el eje x por los valores de la lista periodo.

```python
import matplotlib.pyplot as plt
import numpy as np
juan = [14, 16, 11, 15, 14]
maria = [7, 14, 11, 13, 16]
jose = [11, 9, 15, 5, 17]
periodo = ["primer", "segundo", "tercer", "cuarto", "quinto"]

indx = np.arange(len(periodo))
bar_width = 0.35

fig, ax = plt.subplots()
bar_juan = ax.bar(indx -bar_width/2, juan, bar_width, label="notas de juan")
bar_maria = ax.bar(indx + bar_width/2, maria, bar_width, label="notas de maria")

posiciones = range(len(periodo))
# reemplaza etiqueta de los valores del eje x en este caso 
# posiciones por periodo
plt.xticks(posiciones, periodo)

plt.xlabel("PERIODO")
plt.ylabel("NOTAS")

plt.legend()
plt.show()
```

Diagrama de barras apilado de los tres alumnos (Juan, Maria y Jose)

```python
import matplotlib.pyplot as plt
import numpy as np
juan = [14, 16, 11, 15, 14]
maria = [7, 14, 11, 13, 16]
jose = [11, 9, 15, 5, 17]
periodo = ["primer", "segundo", "tercer", "cuarto", "quinto"]

plt.bar(periodo, juan, label='Juan')
plt.bar(periodo, maria, bottom=juan, label='Maria')
plt.bar(periodo, jose, bottom=np.array(juan)+np.array(maria), label='Jose')

plt.title('Notas por periodo')
plt.xlabel('Periodo')
plt.ylabel('Notas')
plt.legend()

plt.show()
```

Diagrama de barras contiguo de los tres alumnos.

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
juan = [14, 16, 11, 15, 14]
maria = [7, 14, 11, 13, 16]
jose = [11, 9, 15, 5, 17]
periodo = ["primer", "segundo", "tercer", "cuarto", "quinto"]

# Configuración del gráfico
width = 0.25  # Ancho de las barras
x = np.arange(len(periodo))  # Posiciones en el eje x

# Crear el gráfico de barras
fig, ax = plt.subplots()
bar1 = ax.bar(x - width, juan, width, label='Juan')
bar2 = ax.bar(x, maria, width, label='Maria')
bar3 = ax.bar(x + width, jose, width, label='Jose')

# Configurar etiquetas y leyenda
ax.set_xlabel('Periodo')
ax.set_ylabel('Valor')
ax.set_title('Diagrama de Barras Contiguo')
ax.set_xticks(x)
ax.set_xticklabels(periodo)
ax.legend()

# Mostrar el gráfico
plt.show()
```
