```python
"""Del ejemplo anterior 'subplots1.ipynb' vemos que ambos subgraficos 
    el rango de sus ejes y no coinciden, para ello vamos a hacer uso
    del metodo ylim()"""

import matplotlib.pyplot as plt

periodo1, notas_matematicas = range(0, 5), [11, 16, 14, 7, 13]
periodo2, notas_historia = range(0, 5), [8, 12, 15, 14, 11]
alumnos = ["Juan", "Maria", "Luis", "Jorge", "Ana"]

fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1)
ax1.bar(periodo1, notas_matematicas, tick_label = alumnos)
ax1.set_title("matematicas")

"""Se obtiene los valores minimos y maximos del eje y actual (ymin, ymax)
  usando plt.ylim(). Esto se usara mas adelante para asegurarse de que
  ambos subgraficos tengan el mismo rango en el eje y."""
ymin, ymax = plt.ylim()

ax2 = fig.add_subplot(1, 2, 2)
ax2.bar(periodo2, notas_historia, tick_label = alumnos)
ax2.set_title("historia")

"""Se ajusta el rango en el eje y de ax2 para que sea igual al rango
    del primer subgrafico usando plt.ylim(ymin, ymax)"""
plt.ylim(ymin, ymax)

plt.show()
```
