```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

tipos_de_sangre = ["O", "A", "B", "AB"]
num_personas = [212, 84, 45, 19]
hombres = [107, 40, 19, 11]
mujeres = [105,44, 26, 8]

tipos_de_sangre_por_genero = ["hombres_o","mujeres_o","hombres_a","mujeres_a",
                                "hombres_b","mujeres_b","hombres_ab","mujeres_ab"]


fig, ax = plt.subplots()

# la funcion cmap se usa para seleccionar colores especificos de la tableta de colores 'tab20c' de matplotlib
cmap = plt.colormaps["tab20c"]

""" con np.arange(4) generamos un array de numeros enteros desde 0 a 3 osea array([0, 1, 2, 3])
 y con np.arange(4)*5 generamos el array([0, 1, 2, 3]) multiplicado por 5 lo cual nos da array([0, 5, 10, 15]) 
 entonces cmap(np.arange(4)*5) sera cmap(array([0, 5, 10, 15]))
 cmap es una funcion que cuando se le pasa un array de indices devuelve los colores correspondiente de la paleta 
 de colores; en este ejemplo cmap(array([0, 5, 10, 15])) selecciona los colores en los indices 0, 5, 10, 15 de
 la paleta tab20c que contiene 20 colores organizados en grupos de cuatro colores similares
 y usando np.arange(4)*5 nos aseguramos de conseguir una buena separacion de colores dentro de la tableta 'tab20c'"""

color_ext = cmap(np.arange(4)*5)

# de lo explicado anteriormente los indices 1 y 2 seran colores diferentes pero similares, los indices 5 y 6 seran tambien 
# colores diferentes pero similares y asi suscesivamente 
color_int = cmap([1,2,5,6,9,10,13,14])

# Configuracion de la relacion de aspecto para que el grafico sea circular
ax.axis('equal')

# Creamos el grafico de donut exterior, usando desempaquetando de variables 
pie1,por1, _ = ax.pie(num_personas, radius=1.4, labels=tipos_de_sangre, autopct="%1.2f%%", pctdistance=1.3, colors=color_ext) 

# Ajustamos el ancho del primer donut (donut exterior)
plt.setp(pie1, width=0.4, edgecolor='white')

# Creamos el grafico de donut interior, usando desempaquetado de variables
pie2,por2, _ = ax.pie([107, 105, 40, 44, 19, 26, 11, 8], labels=["HO", "MO", "HA", "MA", "HB", "MB", "HAB", "MAB"], autopct="%1.2f%%", colors=color_int )

# Ajustamos el ancho del segundo donut (donut interior)
plt.setp(pie2, width=0.4, edgecolor='white')

# Ajustamos los margenes del grafico
plt.margins(0, 0)

# Generamos las leyendas de los dos donuts
primera_leyenda = plt.legend(pie1,tipos_de_sangre,title="Tipos de Sangre", loc='upper left', bbox_to_anchor=(1,1))
segunda_leyenda = plt.legend(pie2,tipos_de_sangre_por_genero, title="Por Gnero",loc='upper right', bbox_to_anchor=(1.6,1))

# Agregamos la primera leyenda al eje
ax.add_artist(primera_leyenda)

# Ajustamos el tipo de letra del titulo
title_font = {
    'family' : 'serif',
    'color' : 'darkblue',
    'weight' : 'bold',
    'size' : 16,
}

# Agregamos titulo del grafico con espaciado y tipo de letra ajustados
plt.title("Donuts anidados de tipos de sangre por genero",pad=80,fontdict=title_font)

# Mostramos el grafico
plt.show()
```
