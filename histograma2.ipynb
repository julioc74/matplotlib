```python


import matplotlib.pyplot as plt 
import numpy as np 

valores = np.random.randint(1, 1000, size=100)

#Se crea una lista de 10 elementos inicializados en 0, donde cada posici√≥n representa la cantidad de valores en un rango de 100 unidades.
hist = [0] * 10

# En este bucle a medida que se encuentra un valor en un determinado intervalo se incrementa en uno su respectivo elemento contador.
for valor in valores:
    if   valor < 101:  hist[0] += 1 
    elif valor < 201:  hist[1] += 1
    elif valor < 301:  hist[2] += 1
    elif valor < 401:  hist[3] += 1    
    elif valor < 501:  hist[4] += 1
    elif valor < 601:  hist[5] += 1
    elif valor < 701:  hist[6] += 1
    elif valor < 801:  hist[7] += 1
    elif valor < 901:  hist[8] += 1
    elif valor < 1001:  hist[9] += 1
    
x = list(range(1, 11))
etiquetas = ['1-100', '101-200', '201-300', '301-400', '401-500', '501-600', '601-700', '701-800', '801-900', '901-1000']

plt.bar(x, hist, tick_label=etiquetas, width=1)
plt.title("histograma de 100 valores aleatorios entre 1 y 1000")
plt.xlabel("valores")
plt.ylabel("frecuencia")
plt.xticks(rotation=90)
plt.show()


    
```
