```python
import matplotlib.pyplot as plt
import numpy as np

variables = ["velocidad", "tecnica", "fuerza", "inteligencia", "concentracion"]
valores = [8, 4, 7, 9, 5]

# Crea el ángulo para cada variable en radianes
num_var = len(variables)
angulos = np.linspace(0, 2*np.pi, num_var, endpoint=False).tolist()

# Añade el primer valor al final para cerrar la grafica
valores += valores[:1]
angulos += angulos[:1]

# Crea la figura
fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))

# Dibuja el gráfico de radar
ax.fill(angulos, valores, color="blue", alpha=0.25)
ax.plot(angulos, valores, color="blue", linewidth=2)

# Agrega las etiquetas de cada variable en el gráfico
ax.set_xticks(angulos[:-1])
ax.set_xticklabels(variables)

plt.show()
```
